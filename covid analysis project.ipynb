## 6. Date time format :

 ## a. Convert date column in datetime format using 
pandas.to_datetime


```python
import warnings
warnings.simplefilter(action='ignore', category=pd.errors.SettingWithCopyWarning)   #prevent the warning error
def convert_date(date_str):
    try:
        return pd.to_datetime(date_str, format='%d/%m/%y', dayfirst=True)
    except ValueError:
        return pd.to_datetime('01/01/1900', format='%d/%m/%y')
d['date'] = d['date'].apply(convert_date)
d["date"]
```
